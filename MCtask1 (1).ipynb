# Student Performance Analysis

This notebook analyzes student performance using Python.

## Step 1: Import Libraries

We import required libraries for data analysis and visualization.

```python
# ================================
# STEP 1: IMPORT LIBRARIES
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 3: Load Dataset

The dataset is loaded into a Pandas DataFrame.

```python
# ================================
# STEP 3: LOAD DATASET
# ================================
df = pd.read_csv('student-mat.csv')
```

## Step 4: Data Exploration

We check:
- First few rows
- Shape of dataset
- Data types

```python
# ================================
# STEP 4: EXPLORE DATA
# ================================
print("First 5 rows:")
print(df.head())

print("\nShape of dataset:")
print(df.shape)

print("\nData types:")
print(df.dtypes)
```

## Step 5: Missing Values

Check if any values are missing in dataset.

```python
# ================================
# STEP 5: CHECK MISSING VALUES
# ================================
print("\nMissing values:")
print(df.isnull().sum())
```

## Step 6: Remove Duplicates

Duplicate rows are removed to clean data.

```python
# ================================
# STEP 6: REMOVE DUPLICATES
# ================================
df = df.drop_duplicates()
print("\nAfter removing duplicates, shape:")
print(df.shape)
```

## Step 7: Data Analysis

We perform:
- Average grade calculation
- Count of high scoring students
- Correlation analysis
- Gender-wise performance

```python
# STEP 7: ANALYSIS
# ================================

# Average final grade (G3)
avg_grade = df['G3'].mean()
print("\nAverage Final Grade (G3):", avg_grade)

# Students scoring above 15
above_15 = df[df['G3'] > 15].shape[0]
print("Students scoring above 15:", above_15)

# Correlation: study time vs grades
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and G3:", correlation)

# Gender performance
gender_avg = df.groupby('sex')['G3'].mean()
print("\nAverage grades by gender:")
print(gender_avg)
```

## Step 8: Data Visualization

We create:
- Histogram
- Scatter plot
- Bar chart

```python
# ================================
# STEP 8: VISUALIZATION
# ================================

# Histogram of grades
plt.figure()
plt.hist(df['G3'], bins=10)
plt.title("Histogram of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()

# Scatter plot: study time vs grades
plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()

# Bar chart: gender vs average score
plt.figure()
gender_avg.plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
